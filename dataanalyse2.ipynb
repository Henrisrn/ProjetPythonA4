Traitement de la donnée

```python


# Fonctions de traduction pour chaque colonne
def translate_sex(value):
    return 'Masculin' if value == 1 else 'Féminin' if value == 0 else 'Inconnu'

def translate_chest_pain_type(value):
    types = {
        1: "Douleur typique de l'angine",
        2: "Douleur atypique de l'angine",
        3: 'Douleur non angineuse',
        4: 'Asymptomatique'
    }
    return types.get(value, 'Inconnu')

def translate_fasting_blood_sugar(value):
    return 'Supérieur à 120 mg/dl' if value == 1 else 'Inférieur ou égal à 120 mg/dl' if value == 0 else 'Inconnu'

def translate_resting_ecg(value):
    results = {
        0: 'Normal',
        1: "Anomalie de l'onde ST-T",
        2: 'Hypertrophie ventriculaire probable ou certaine'
    }
    return results.get(value, 'Inconnu')

def translate_exercise_induced_angina(value):
    return 'Oui' if value == 1 else 'Non' if value == 0 else 'Inconnu'

def translate_thal(value):
    thal_types = {
        3: 'Normal',
        6: 'Défaut fixe',
        7: 'Défaut réversible'
    }
    return thal_types.get(value, 'Inconnu')

def translate_disease_presence(value):
    if value == 0:
        return 'Absence de maladie'
    elif 1 <= value <= 4:
        return 'Présence de maladie'
    else:
        return 'Inconnu'

# Application des fonctions de traduction à chaque colonne
data['sex_translation'] = data['sex'].map(translate_sex)
data['chest_pain_type_translation'] = data['chest_pain_type'].map(translate_chest_pain_type)
data['fasting_blood_sugar_translation'] = data['Glycemie_a_jeun'].map(translate_fasting_blood_sugar)
data['resting_ecg_translation'] = data['ecg_results'].map(translate_resting_ecg)
data['exercise_induced_angina_translation'] = data['angine_poitrine'].map(translate_exercise_induced_angina)
data['thal_translation'] = data['Thalassemie'].map(translate_thal)
data['disease_presence_translation'] = data['disease_presence'].map(translate_disease_presence)

# Vous pouvez répéter cette méthode pour les autres colonnes selon le besoin.
```
